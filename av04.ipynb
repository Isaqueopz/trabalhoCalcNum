1. Usando o método dos mínimos quadrados, calcule a equação da reta que melhor ajusta os pontos. Exiba o gráfico da função e os pontos.

$$
\begin{array}{c|cccc}
x & -1 & 0 & 1 & 2 \\
\hline
y & 0 & -1 & 0 & 7
\end{array}
$$

obs.: Cálculo "manual" usando Python como calculadora.

```python
# Questão 1: Ajuste de reta por mínimos quadrados
# Pontos fornecidos
import numpy as np
import matplotlib.pyplot as plt

# Dados
x = np.array([-1, 0, 1, 2])
y = np.array([0, -1, 0, 7])

# Explicação do procedimento
print("Vamos calcular a equação da reta y = a*x + b que melhor ajusta os pontos pelo método dos mínimos quadrados.")

# Montando o sistema normal
n = len(x)
Sx = np.sum(x)
Sy = np.sum(y)
Sxx = np.sum(x**2)
Sxy = np.sum(x*y)

# Sistema:
# a*Sxx + b*Sx = Sxy
# a*Sx  + b*n  = Sy
A = np.array([[Sxx, Sx], [Sx, n]])
B = np.array([Sxy, Sy])

# Resolvendo o sistema
sol = np.linalg.solve(A, B)
a, b = sol
print(f"Coeficientes encontrados: a = {a:.2f}, b = {b:.2f}")
print(f"Equação da reta ajustada: y = {a:.2f}x + {b:.2f}")

# Gráfico
plt.scatter(x, y, color='red', label='Pontos dados')
x_plot = np.linspace(min(x)-0.5, max(x)+0.5, 100)
y_plot = a*x_plot + b
plt.plot(x_plot, y_plot, label='Reta ajustada')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ajuste de reta por mínimos quadrados')
plt.legend()
plt.grid(True)
plt.show()
```

2. Construa o polinômio de interpolação, na forma de Lagrange, para a função $y=\sin \pi x$, escolhendo os pontos $x_0 = 0, x_1= \dfrac{1}{6}$ e $x_2 = \dfrac{1}{2}$.

obs.: Cálculo "manual" usando Python como calculadora.

```python
# Questão 2: Polinômio de Lagrange para y = sin(pi*x)
import numpy as np
from sympy import symbols, expand, simplify, lambdify, pi, sin
import matplotlib.pyplot as plt

# Pontos escolhidos
x0 = 0
x1 = 1/6
x2 = 1/2
y0 = np.sin(np.pi * x0)
y1 = np.sin(np.pi * x1)
y2 = np.sin(np.pi * x2)

print(f"Pontos usados: (0, {y0:.3f}), (1/6, {y1:.3f}), (1/2, {y2:.3f})")

# Construindo o polinômio de Lagrange manualmente
x = symbols('x')
L0 = ((x - x1)*(x - x2))/((x0 - x1)*(x0 - x2))
L1 = ((x - x0)*(x - x2))/((x1 - x0)*(x1 - x2))
L2 = ((x - x0)*(x - x1))/((x2 - x0)*(x2 - x1))
P = y0*L0 + y1*L1 + y2*L2
P = expand(P)
print("Polinômio de Lagrange (expandido):")
print(P)

# Gráfico
f_real = lambdify(x, sin(pi*x), 'numpy')
f_lagrange = lambdify(x, P, 'numpy')
x_vals = np.linspace(0, 0.5, 100)
plt.plot(x_vals, f_real(x_vals), label='y = sin(pi*x)')
plt.plot(x_vals, f_lagrange(x_vals), '--', label='Polinômio de Lagrange')
plt.scatter([x0, x1, x2], [y0, y1, y2], color='red', label='Pontos')
plt.legend()
plt.title('Interpolação de Lagrange para y = sin(pi*x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

4. Selecione dois métodos de interpolação de pontos, utilizando as funções disponíveis em Python.

    (a) Fixe uma semente aleatória e gere um conjunto de $n$ pontos $(x_i, y_i)$;

    (b) Realize uma análise comparativa entre os dois métodos escolhidos. Há diferenças no comportamento dos métodos quando o número de pontos é significativamente aumentado? Se sim, exiba este caso.

```python
# Questão 4: Comparação de métodos de interpolação
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange, CubicSpline

# (a) Gerando pontos aleatórios
np.random.seed(42)  # Semente fixa
n = 10
x = np.sort(np.random.rand(n))
y = np.sin(2 * np.pi * x) + np.random.uniform(-0.1, 0.1, n)

# (b) Interpolação de Lagrange e Spline Cúbico
lagrange_poly = lagrange(x, y)
spline = CubicSpline(x, y)

# Avaliação
x_plot = np.linspace(0, 1, 200)
y_lagrange = lagrange_poly(x_plot)
y_spline = spline(x_plot)
y_true = np.sin(2 * np.pi * x_plot)

plt.figure(figsize=(10,5))
plt.plot(x_plot, y_true, label='Função original', color='black', linestyle=':')
plt.plot(x_plot, y_lagrange, label='Interpolação de Lagrange')
plt.plot(x_plot, y_spline, label='Spline Cúbico')
plt.scatter(x, y, color='red', label='Pontos')
plt.legend()
plt.title('Comparação de métodos de interpolação')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

# Análise para n grande
def compara_n_grande():
    n2 = 30
    x2 = np.sort(np.random.rand(n2))
    y2 = np.sin(2 * np.pi * x2) + np.random.uniform(-0.1, 0.1, n2)
    lagrange_poly2 = lagrange(x2, y2)
    spline2 = CubicSpline(x2, y2)
    x_plot2 = np.linspace(0, 1, 400)
    plt.figure(figsize=(10,5))
    plt.plot(x_plot2, np.sin(2 * np.pi * x_plot2), label='Função original', color='black', linestyle=':')
    plt.plot(x_plot2, lagrange_poly2(x_plot2), label='Lagrange (n=30)')
    plt.plot(x_plot2, spline2(x_plot2), label='Spline Cúbico (n=30)')
    plt.scatter(x2, y2, color='red', label='Pontos')
    plt.legend()
    plt.title('Comportamento para n grande')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.show()
    print('Note que o polinômio de Lagrange pode oscilar muito para n grande (fenômeno de Runge), enquanto o spline mantém suavidade.')

compara_n_grande()
```

### Análise comparativa dos métodos de interpolação

Observa-se que a interpolação de Lagrange, por ser um polinômio único de grau elevado quando o número de pontos aumenta, pode apresentar oscilações acentuadas entre os pontos (fenômeno de Runge), especialmente nas extremidades do intervalo. Isso é visível nos gráficos quando aumentamos o número de pontos: o polinômio de Lagrange pode se afastar bastante da função original e apresentar grandes variações.

Comparando os métodos, nota-se que o Spline Cúbico constrói funções polinomiais de grau 3 em cada subintervalo, garantindo suavidade e continuidade das derivadas. Por isso, mesmo com muitos pontos, o spline mantém estabilidade e aproxima melhor a função original, sem oscilações exageradas.

À medida que o número de pontos aumenta, o spline cúbico se mostra mais estável e confiável, enquanto o polinômio de Lagrange pode se tornar inadequado devido às oscilações. Para poucos pontos ambos os métodos podem funcionar bem, mas para muitos pontos o spline cúbico é preferível.

5. Ajuste de curvas por Mínimos Quadrados

    A. Geração dos dados
    - Fixe uma semente aleatória;
    - Gere $n$ valores $x_i$ em um intervalo a sua escolha (usando np.linspace)
    - Para a geração dos valores $y_i$, considere a função $$y_i = x_i^2 + \varepsilon_i$$    
        A expressão $\varepsilon_i$ representa pequenas variações ou erros nos dados. Neste caso, use 

        $\varepsilon_i=$ `np.random.uniform(-0.5, 0.5, len(x))`


    B. Ajuste de modelos

    Utilizando funções de ajuste de curvas disponíveis em Python:

    1. Ajuste um modelo linear aos dados;

    2. Ajuste um modelo polinomial de grau 2;

    3. Ajuste um modelo polinomial de grau maior (por exemplo, grau 4 ou 5).
    

    C. Análise

    1. Compare visualmente os ajustes obtidos;

    2. Compare os erros dos modelos. Uma forma de fazer isso é calculando o resíduo (ver notas de aula). Mas aqui sugiro o uso de uma das chamadas medidas globais de erro: o Erro Quadatico Médio (MSE).

    3. Discuta o comportamento dos modelos à medida que o grau do polinômio aumenta.

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.polynomial import Polynomial

# A. Geração dos dados
np.random.seed(42)
n = 20
x = np.linspace(-2, 2, n)
epsilon = np.random.uniform(-0.5, 0.5, len(x))
y = x**2 + epsilon

# B. Ajuste de modelos
# 1. Linear
p1 = np.polyfit(x, y, 1)
y_pred1 = np.polyval(p1, x)

# 2. Polinomial grau 2
g2 = np.polyfit(x, y, 2)
y_pred2 = np.polyval(g2, x)

# 3. Polinomial grau 5
g5 = np.polyfit(x, y, 5)
y_pred5 = np.polyval(g5, x)

# C. Análise visual
t = np.argsort(x)
plt.figure(figsize=(10,6))
plt.scatter(x, y, color='black', label='Dados')
plt.plot(x[t], y_pred1[t], label='Linear', linestyle='--')
plt.plot(x[t], y_pred2[t], label='Polinomial grau 2')
plt.plot(x[t], y_pred5[t], label='Polinomial grau 5')
plt.legend()
plt.title('Ajuste de curvas por mínimos quadrados')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()

# C.2. Erros (MSE calculado manualmente com NumPy)
mse1 = np.mean((y - y_pred1)**2)
mse2 = np.mean((y - y_pred2)**2)
mse5 = np.mean((y - y_pred5)**2)

print(f"MSE Linear: {mse1:.4f}")
print(f"MSE Grau 2: {mse2:.4f}")
print(f"MSE Grau 5: {mse5:.4f}")

# C.3. Discussão
discussao = '''\nÀ medida que o grau do polinômio aumenta, o modelo pode se ajustar melhor aos dados, mas corre o risco de sobreajuste (overfitting), capturando o ruído ao invés da tendência real. O modelo linear não representa bem a tendência quadrática dos dados. O polinômio de grau 2 é o mais adequado, pois reflete a função original. O polinômio de grau 5 pode oscilar demais fora dos pontos de dados.\n'''
print(discussao)
```

### Comparação visual e análise dos modelos de ajuste

O gráfico acima mostra os dados originais (com ruído) e as curvas ajustadas pelos três modelos: linear, polinomial de grau 2 e polinomial de grau 5. O modelo linear não consegue capturar a curvatura dos dados, resultando em um erro maior. O polinômio de grau 2 se ajusta muito bem, pois reflete a função original $y = x^2$. Já o polinômio de grau 5, apesar de apresentar um erro baixo nos pontos, começa a oscilar e a se ajustar ao ruído dos dados (overfitting), especialmente nas extremidades do intervalo.

Os valores do Erro Quadrático Médio (MSE) para cada modelo são impressos acima e permitem comparar quantitativamente a qualidade do ajuste. Modelos de grau muito alto podem parecer melhores nos pontos amostrados, mas tendem a capturar o ruído e não a tendência real dos dados.
